```python
import sarracen
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Render column density of gas and dust for $t/t_{\rm c} = 2, 3, 4, 5$

```python
sdf_2f_40 = sarracen.read_phantom('2fluid/2f3mic256_00040')
sdf_2f_60 = sarracen.read_phantom('2fluid/2f3mic256_00060')
sdf_2f_80 = sarracen.read_phantom('2fluid/2f3mic256_00080')
sdf_2f_100 = sarracen.read_phantom('2fluid/2f3mic256_00100')
```

```python
sdf_1f_40 = sarracen.read_phantom('limited/3mic256l_00040')
sdf_1f_60 = sarracen.read_phantom('limited/3mic256l_00060')
sdf_1f_80 = sarracen.read_phantom('limited/3mic256l_00080')
sdf_1f_100 = sarracen.read_phantom('limited/3mic256l_00100')
```

```python
sdf_1f_40.calc_density() 
sdf_1f_40['rho_d'] = sdf_1f_40['rho'] * sdf_1f_40['dustfrac']
sdf_1f_40['rho_g'] = sdf_1f_40['rho'] * (1.0 - sdf_1f_40['dustfrac'])

sdf_1f_60.calc_density() 
sdf_1f_60['rho_d'] = sdf_1f_60['rho'] * sdf_1f_60['dustfrac']
sdf_1f_60['rho_g'] = sdf_1f_60['rho'] * (1.0 - sdf_1f_60['dustfrac'])

sdf_1f_80.calc_density() 
sdf_1f_80['rho_d'] = sdf_1f_80['rho'] * sdf_1f_80['dustfrac']
sdf_1f_80['rho_g'] = sdf_1f_80['rho'] * (1.0 - sdf_1f_80['dustfrac'])

sdf_1f_100.calc_density() 
sdf_1f_100['rho_d'] = sdf_1f_100['rho'] * sdf_1f_100['dustfrac']
sdf_1f_100['rho_g'] = sdf_1f_100['rho'] * (1.0 - sdf_1f_100['dustfrac'])
```

```python
#gas-10mic 40-100 (2-fluis and 1-fluid)


positions = [
    [0.05, 0.55, 0.28, 0.28],  # Top-left subplot
    [0.27, 0.55, 0.28, 0.28],  # Top-center-left subplot
    [0.49, 0.55, 0.28, 0.28],  # Top-center-right subplot
    [0.71, 0.55, 0.28, 0.28],  # Top-right subplot
    [0.05, 0.20, 0.28, 0.28],  # Bottom-left subplot
    [0.27, 0.20, 0.28, 0.28],  # Bottom-center-left subplot
    [0.49, 0.20, 0.28, 0.28],  # Bottom-center-right subplot
    [0.71, 0.20, 0.28, 0.28]   # Bottom-right subplot
]

fig = plt.figure()

axes = []
for pos in positions:
    axes.append(fig.add_axes(pos))

    
#Top row
sdf_2f_40[sdf_2f_40.itype == 1].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', vmin=0.1, vmax=10, ax=axes[0])
axes[0].set_xticklabels([])
axes[0].set_xlabel('')
# axes[0].set_title('2-fluid method')
axes[0].text(0.05, 0.85, r'$\mathbf{2t_c}$', fontsize = 12, color='w', weight='bold')
axes[0].tick_params(axis = 'y', labelsize = 8)

sdf_2f_60[sdf_2f_60.itype == 1].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', vmin=0.1, vmax=10, ax=axes[1])
axes[1].set_xticklabels([])
axes[1].set_xlabel('')
axes[1].set_yticklabels([])
axes[1].set_ylabel('')
axes[1].text(0.05, 0.85, r'$\mathbf{3t_c}$', fontsize = 12, color='w', weight='bold')


sdf_2f_80[sdf_2f_80.itype == 1].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', vmin=0.1, vmax=10, ax=axes[2])
axes[2].set_xticklabels([])
axes[2].set_xlabel('')
axes[2].set_yticklabels([])
axes[2].set_ylabel('')
axes[2].text(0.05, 0.85, r'$\mathbf{4t_c}$', fontsize = 12, color='w',weight='bold')


sdf_2f_100[sdf_2f_100.itype == 1].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', vmin=0.1, vmax=10, ax=axes[3])
axes[3].set_xticklabels([])
axes[3].set_xlabel('')
axes[3].set_yticklabels([])
axes[3].set_ylabel('')
axes[3].text(0.05, 0.85, r'$\mathbf{5t_c}$', fontsize = 12, color='w', weight='bold')


#bottom row
sdf_1f_40.render('rho_g', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', dens_weight=False,  vmin=0.1, vmax=10, ax=axes[4])
# axes[4].set_title('1-fluid method')
axes[4].tick_params(axis = 'x', labelsize = 8)
axes[4].tick_params(axis = 'y', labelsize = 8)

sdf_1f_60.render('rho_g', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', dens_weight=False, vmin=0.1, vmax=10, ax=axes[5])
axes[5].set_yticklabels([])
axes[5].set_ylabel('')
axes[5].tick_params(axis = 'x', labelsize = 8)

sdf_1f_80.render('rho_g', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', dens_weight=False, vmin=0.1, vmax=10, ax=axes[6])
axes[6].set_yticklabels([])
axes[6].set_ylabel('')
axes[6].tick_params(axis = 'x', labelsize = 8)

sdf_1f_100.render('rho_g', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', dens_weight=False, vmin=0.1, vmax=10, ax=axes[7])
axes[7].set_yticklabels([])
axes[7].set_ylabel('')
axes[7].tick_params(axis = 'x', labelsize = 8)




fig.text(0.42, 0.85, '2-fluid method', fontsize = 12)
fig.text(0.42, 0.50, '1-fluid method', fontsize = 12)
fig.supylabel('$3\mu m$')
# fig.savefig('3mic_gas40_100.pdf')
```

# Render column density of dust and gas together for $t/t_{\rm c} = 2 $

```python
fig = plt.figure(figsize=(8, 6))

# Define the positions of the subplots: [left, bottom, width, height]
positions = [
    [0.1, 0.55, 0.4, 0.4],  # Top-left subplot
    [0.42, 0.55, 0.4, 0.4],  # Top-right subplot
    [0.1, 0.12, 0.4, 0.4],    # Bottom-left subplot
    [0.42, 0.12, 0.4, 0.4]    # Bottom-right subplot
]

# Add subplots using add_axes
axes = []
for pos in positions:
    axes.append(fig.add_axes(pos))

# Rendering the subplots
sdf_2f_40[sdf_2f_40.itype == 1].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='gist_heat', vmin=0.1, vmax=10, ax=axes[0])
axes[0].set_title('2-fluid method')
axes[0].set_xticks([])
axes[0].set_xlabel('')


sdf_1f_40.render('rho_g', log_scale=True, backend='cpu', cmap='gist_heat',cbar=False, vmin=0.1, vmax=10, dens_weight=False, ax=axes[1])
axes[1].set_title('1-fluid method')
axes[1].set_xticks([])
axes[1].set_xlabel('')
axes[1].set_yticks([])
axes[1].set_ylabel('')

sdf_2f_40[sdf_2f_40.itype == 7].render('rho', log_scale=True, backend='cpu', cbar=False, cmap='cividis', vmin=0.001, vmax=0.1, ax=axes[2])
# axes[2].set_title('Dust in 2-fluid method')

sdf_1f_40.render('rho_d', log_scale=True, backend='cpu', cmap='cividis',cbar=False, vmin=0.001, vmax=0.1, dens_weight=False, ax=axes[3])
# axes[3].set_title('Dust in 1-fluid method')
axes[3].set_yticks([])
axes[3].set_ylabel('')

fig.tight_layout()
fig.suptitle(r'3 $\mu$m', fontsize=16, x = 0.46 , y=1)
# fig.savefig('3mic_100th.pdf')

```

#  Slices of gas and dust in the two methods for $t/t_{\rm c}=5$
